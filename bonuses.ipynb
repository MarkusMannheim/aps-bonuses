```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

sns.set_style("darkgrid")
sns.set_theme()
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 3.5})

data = pd.read_excel("./bonuses.xlsx")
grouped_data = data.groupby("year").sum()
grouped_data["ratio"] = grouped_data["paid"] / grouped_data["employees"]

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1, 1, 1)
grouped_data["ratio"].plot()
plt.title("SES staff who received performance bonuses", fontdict={"size": 18})
plt.xlabel(None)
plt.ylabel(None)
ax.yaxis.set_major_formatter(ticker.PercentFormatter(1, decimals=0))
plt.savefig("ses_bonuses.png")
```
